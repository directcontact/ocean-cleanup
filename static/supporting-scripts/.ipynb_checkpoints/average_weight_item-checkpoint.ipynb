```python
#import modules
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read csv
df1 = pd.read_csv('data\Data_Level5_BAH_OceanCleanup.final_version.csv', low_memory = False)
```

```python
#creates subset and initialize variables
subset = df1.loc[:, '# of bags':'Plastic Pieces']
df1['mean'] = ''
df2 = pd.DataFrame()
```

```python
#for each row, if the column doesnt equals to 0, then add to the list to get the average weight of each column in that row
for i, row in df1.iterrows():
    c = []
    for col in subset.columns:
        if df1.iloc[i][col] != 0:       
            c.append(col)
            number = df1.iloc[i]['Pounds']/len(c)
    df1.at[i,'mean'] = number
```

```python
#if the values of series are str, change them to int
for i in range(13,60):
    if type(df1[df1.columns[i]][0]) == str:
        df1[f'{df1.columns[i]}']=df1[f'{df1.columns[i]}'].astype('int64')
```

```python
#create new columns in df2 with the mean divided by the original values
for i in range(13,60):
    name = df1.columns[i]
    df2[f'{name}_mean']=df1['mean']/df1[df1.columns[i]]
```

```python
#replacing inf with nan
df2=df2.replace([np.inf, -np.inf], np.nan)
df2
```

```python
#get the median of each column in df2 and put it into a list
med_lst=[]
for i in range(len(df2.columns)):
    med = np.nanmedian(df2[df2.columns[i]], axis=0)
    med_lst.append(med)
```

```python
#creating a dataframe to store the results
df3 = pd.DataFrame()
df3['Garbage Item'] = ['Cigarette Butts', 'Food Wrappers (candy, chips, etc.)',
       'Take Out/Away Containers (Plastic)', 'Take Out/Away Containers (Foam)',
       'Bottle Caps (Plastic)', 'Bottle Caps (Metal)', 'Lids (Plastic)',
       'Straws, Stirrers', 'Forks, Knives, Spoons',
       'Beverage Bottles (Plastic)', 'Beverage Bottles (Glass)',
       'Beverage Cans', 'Grocery Bags (Plastic)', 'Other Plastic Bags',
       'Paper Bags', 'Cups, Plates (Paper)', 'Cups, Plates (Plastic)',
       'Cups, Plates (Foam)', 'Fishing Buoys, Pots & Traps',
       'Fishing Net & Pieces', 'Fishing Line (1 yard/meter = 1 piece)',
       'Rope (1 yard/meter = 1 piece)', 'Fishing Gear (Clean Swell)',
       '6-Pack Holders', 'Other Plastic/Foam Packaging',
       'Other Plastic Bottles (oil, bleach, etc.)', 'Strapping Bands',
       'Tobacco Packaging/Wrap', 'Other Packaging (Clean Swell)',
       'Appliances (refrigerators, washers, etc.)', 'Balloons', 'Cigar Tips',
       'Cigarette Lighters', 'Construction Materials', 'Fireworks', 'Tires',
       'Toys', 'Other Trash (Clean Swell)', 'Condoms', 'Diapers', 'Syringes',
       'Tampons/Tampon Applicators', 'Personal Hygiene (Clean Swell)',
       'Foam Pieces', 'Glass Pieces', 'Plastic Pieces', '# of bags']
```

```python
#creating a dataframe for the weights
df4 = pd.DataFrame(med_lst, columns = ['Average Weight (in Pounds)'])
```

```python
#combining both dataframes
df3['Average Weight (in Pounds)'] = df4['Average Weight (in Pounds)']
```
